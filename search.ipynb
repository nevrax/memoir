```python
import pandas as pd

pd.options.display.float_format = '{:.8f}'.format

filter = 'MANA'

header = [
    'Date(UTC)', 'Pair', 'Side',
    'Price', 'ExecVal', 'AmVal'
]


df = pd.read_csv('aggregated-step2.csv')
fdf = df[df['Pair'].str.startswith(filter)]

#print(fdf.head(3))
#print(fdf[header].head(5).to_string(index=False, decimal=','))

fdf[header].head(3)
```

```python
#import pandas as pd

#df = fdf


df
f
total_bought = 0.0
total_cost = 0.0
pnl_list = []

for index, row in df.iterrows():
    if row['Side'] == 'BUY':
        # Update the total amount bought and total cost
        total_bought += row['ExecVal']
        total_cost += row['ExecVal'] * row['Price']
    elif row['Side'] == 'SELL':
        if total_bought > 0:
            # Calculate the average buy price
            avg_buy_price = total_cost / total_bought
            
            # Calculate PnL for the sell transaction
            pnl = (row['Price'] - avg_buy_price) * row['ExecVal']
            pnl_list.append(pnl)
            
            # Update the total amount bought and total cost
            total_bought -= row['ExecVal']
            total_cost -= avg_buy_price * row['ExecVal']
        else:
            # If no buys left to match the sell, PnL is considered zero or an error
            pnl_list.append(0.0)
    else:
        pnl_list.append(0.0)

#pnl_list

df['PnL'] = pnl_list
#df
```
